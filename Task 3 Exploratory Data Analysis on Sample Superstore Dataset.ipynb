### Numerical Variable Visualization

```python
plt.figure(figsize=(20,11))
ax=sns.distplot(data["Profit"])
ax.set_yscale('log')
```

As we can see the profit is higher when the product rate is less.

```python
plt.figure(figsize=(20,10))
ax=sns.distplot(data["Sales"])
ax.set_yscale('log')
```

As we can see the sales are higher in the range of 0 - 5000.

```python
plt.figure(figsize=(20,12))
data_group=data.groupby(["State"])["Profit"].sum()
data_group.sort_values(ascending=False,inplace=True)
data_group.plot.bar()
plt.show()
```

As we can see the state California Shows the highest profit and states like New York , Washington, etc. shows good profit.
On the other hand few states like Texas, Ohio has the reverse impact on the profit though the paticipation was more in these states still the states has shown loss than profit.

Checking Shipmode variable with the profit

```python
plt.figure(figsize=(20,10))
data_group_city=data.groupby(["Ship Mode"])["Profit"].sum()
data_group_city.sort_values(ascending=False,inplace=True)
data_group_city.plot.bar()
plt.show()
```

The Standard Class shipping mode makes more profit than others.

Checking City variable with the profit

```python
plt.figure(figsize=(20,10))
group_city=data.groupby(["City"])["Profit"].sum()
data_top=group_city[group_city>5000]
data_top.sort_values(ascending=False,inplace=True)
data_top.plot.bar()
plt.show()
```

As we can see cities like New york city, Los Angeles, Seattle, etc. makes more profit than other cities.

Checking on the city variable with the sales

```python
plt.figure(figsize=(20,10))
group_city=data.groupby(["City"])["Sales"].sum()
top_sales=group_city[group_city>20000]
top_sales.sort_values(ascending=False,inplace=True)
top_sales.plot.bar()
plt.show()
```

As we can see Cities like New York City ,Los Angeles and Seattle had good sales rate and good profit too.

visualizing the Category variable with the sales and region

* The furniture have a good business in the western region and so is the technology
* The office supply in the east is not much as the furniture in the east and the technology
* The technology makes good profit in the south region

```python
plt.figure(figsize=(20, 9))
ax=sns.boxplot(x = 'Segment', y = 'Sales', hue = 'Region', data = data)
ax.set_yscale('log')
plt.show()
```

* The Corporate segment in the west makes good sales.
* The home office in thr Central also has a good profit making business
* The consumers in all the regions are making a good profit
